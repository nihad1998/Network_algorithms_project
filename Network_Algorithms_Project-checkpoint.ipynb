```python
import networkx as nx 
import pandas as pd
import io
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from itertools import product
```

```python
df = pd.read_csv('~/Desktop/Network_project')
```

```python
#Firstly we need to divide our tt columns to 2 columns called x_coor and y_coor
df[['x_coor','y_coor']] = df.tt.str.split(",",expand=True,)
```

```python
#deleting useless columns
df = df.drop(["TYPE", "ville", "tt", "adresse", "code"], axis=1)
```

```python
#Rename fist column (Nom restaurant)
df.rename(columns = {'Nom restaurant ':'Names'}, inplace =True)
```

```python
#df = df.drop(35, axis=0)
```

```python
df['x_coor'] = df['x_coor'].astype(float)
df['y_coor'] = df['y_coor'].astype(float)

print(df)
```

```python
x = df.iloc[:,[1,2]].values
kmeans = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
#Visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], c = 'red', label = 'A')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], c = 'blue', label = 'B')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], c = 'green', label = 'C')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], c = 'black', label = 'D')
#Plotting the centroids of the clusters
#plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
```

```python
car_A = pd.DataFrame(columns=['Names','x_coor', 'y_coor'])
car_B = pd.DataFrame(columns=['Names','x_coor', 'y_coor'])
car_C = pd.DataFrame(columns=['Names','x_coor', 'y_coor'])
car_D = pd.DataFrame(columns=['Names','x_coor', 'y_coor'])
```

```python
car_A
```

```python
j=0
for i in y_kmeans:
  if i==0:
    car_A=car_A.append(df.iloc[j,:])
  if i==1:
    car_B=car_B.append(df.iloc[j,:])
  if i==2:
    car_C=car_C.append(df.iloc[j,:])
  if i==3:
    car_D=car_D.append(df.iloc[j,:])
  j+=1
```

```python
# From there we don't need coordinates since we used them for K_Means, delete them
car_A = car_A.drop(["x_coor","y_coor"], axis=1)
```

```python
#Our indexes like [1,2,5,6] we should do it [0,1,2,3,4]
def reIndecColumns(df):
  df = df.reset_index(drop=True)
  return df

car_A = reIndecColumns(car_A)
car_B = reIndecColumns(car_B)
car_C = reIndecColumns(car_C)
car_D = reIndecColumns(car_D)
car_D
```

```python
#BURDA COMPLETE GRAPH TEMASINI ELEMEK LAZIMDI, BUTUN NODELERDEN BUTUN NODELERE(4 subdata ucun de)
def MakeCompleteGraph(df):
  df_pairs = pd.DataFrame(
      filter(lambda x: x[0]!=x[1], product(df.Names, df.Names)), 
      columns=['Names', 'dest']
  )
  return df_pairs
```

```python
#generate random numbers for weight and distances
#weight shows traffic scale in roads
def Generate_distance_and_Weight(df):
  df = MakeCompleteGraph(df)
  weightList = []
  distanceList = []
  for i in range(len(df)):
    weight = random.randrange(0,10)
    weightList.append(weight)
    distance = random.randrange(100,500)
    distanceList.append(distance)
#Adding new columns to dataframe
  df["Weight"] = weightList
  df["Distance"] = distanceList
  return df
```

```python
car_A = Generate_distance_and_Weight(car_A)
car_B = Generate_distance_and_Weight(car_B)
car_C = Generate_distance_and_Weight(car_C)
car_D = Generate_distance_and_Weight(car_D)
```

```python
print('-------------------CAR A-----------\n')
print(car_A)
print('-------------------CAR B------------\n')
print(car_B)
print('-------------------CAR C------------\n')
print(car_C)
print('-------------------CAR D------------\n')
print(car_D)
```

```python
graph_car_A = nx.from_pandas_edgelist(car_A, source='Names', target='dest' , edge_attr='Weight')
graph_car_B = nx.from_pandas_edgelist(car_B, source='Names', target='dest' , edge_attr='Weight')
graph_car_C = nx.from_pandas_edgelist(car_C, source='Names', target='dest' , edge_attr='Weight')
graph_car_D = nx.from_pandas_edgelist(car_D, source='Names', target='dest' , edge_attr='Weight')
```

```python

plt.figure(figsize=(10,10))
plt.subplot(2,2 , 1)
nx.draw_networkx(graph_car_A, with_labels=True)
plt.subplot(2,2 , 2)
nx.draw_networkx(graph_car_B, with_labels=True)
plt.subplot(2,2 , 3)
nx.draw_networkx(graph_car_C, with_labels=True)
plt.subplot(2,2 , 4)
nx.draw_networkx(graph_car_D, with_labels=True)

plt.show()
```

```python
#minimum spanning tree of four graph based on traffic weight 

plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
A_spanning_tree=nx.minimum_spanning_tree(graph_car_A,algorithm='prim', weight='Weight' )
nx.draw_networkx(A_spanning_tree)

plt.subplot(1,4,2)
B_spanning_tree=nx.minimum_spanning_tree(graph_car_B,algorithm='prim', weight='Weight' )
nx.draw_networkx(B_spanning_tree)

plt.subplot(1,4,3)
C_spanning_tree=nx.minimum_spanning_tree(graph_car_C,algorithm='prim', weight='Weight' )
nx.draw_networkx(A_spanning_tree)

plt.subplot(1,4,4)
D_spanning_tree=nx.minimum_spanning_tree(graph_car_D,algorithm='prim', weight='Weight' )
nx.draw_networkx(A_spanning_tree)

plt.show()
```

```python
#visualising the graph with specific edges colored 

plt.figure(figsize=(25,5))

plt.subplot(1,4,1)
pos = nx.spring_layout(graph_car_A)

nx.draw_networkx_nodes(graph_car_A,pos, with_labels=True)
nx.draw_networkx_edges(graph_car_A, pos, width=0.3 )
#nx.draw_networkx_edges(graph_car_A, pos, edgelist=graph_car_A.edges(),width=4)
nx.draw_networkx_edges(graph_car_A, pos,edgelist=A_spanning_tree.edges(), edge_color='b',width=1.5)
nx.draw_networkx_labels(graph_car_A, pos)

plt.subplot(1,4,2)
pos = nx.spring_layout(graph_car_B)
nx.draw_networkx_nodes(graph_car_B,pos, with_labels=True)
nx.draw_networkx_edges(graph_car_B, pos, width=0.2)
nx.draw_networkx_edges(graph_car_B, pos,edgelist=B_spanning_tree.edges(), width=1.5,edge_color='b')
nx.draw_networkx_labels(graph_car_B, pos)

plt.subplot(1,4,3)

pos = nx.spring_layout(graph_car_C)
nx.draw_networkx_nodes(graph_car_C, pos, with_labels=True)
nx.draw_networkx_edges(graph_car_C, pos, width=0.2)
nx.draw_networkx_edges(graph_car_C, pos,edgelist=C_spanning_tree.edges(), width=1.5,edge_color='b')
nx.draw_networkx_labels(graph_car_C, pos)

plt.subplot(1,4,4)
pos = nx.spring_layout(graph_car_D)
nx.draw_networkx_nodes(graph_car_D, pos , with_labels=True)
nx.draw_networkx_edges(graph_car_D, pos , width=0.2)
nx.draw_networkx_edges(graph_car_D, pos , edgelist=D_spanning_tree.edges(),width=1.5, edge_color='b')
nx.draw_networkx_labels(graph_car_D, pos)
```
